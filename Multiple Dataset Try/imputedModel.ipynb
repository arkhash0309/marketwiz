```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.impute import SimpleImputer
```

```python
sales_df = pd.read_csv("Sales_Data.csv")
sales_df
```

```python
data = {'Day': [1, 2, 3, 4, 5],
        'Month': [1, 2, 3, 4, 5],
        'Year': [2021, 2021, 2021, 2021, 2021],
        'DayOfWeek': ['Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'],
        'DayType': ['Weekday', 'Weekday', 'Weekday', 'Weekend', 'Weekend'],
        'Rainfall': [5.2, 3.1, None, 2.5, 4.0],
        'Week_of_Year': [1, 2, 3, 4, 5],
        'Category': ['YOGHURT', 'BUTTER', 'CHEESE', 'COFFEE', 'FLOUR']}

df = pd.DataFrame(data)

x = df.drop('Category', axis = 1)
y = df['Category']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Split the data into training and testing components
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

# Create SimpleImputer with strategy='mean' for Rainfall column
imputer = SimpleImputer(strategy='mean')

# Impute missing values in the training and testing sets
X_train['Rainfall'] = imputer.fit_transform(X_train[['Rainfall']])
X_test['Rainfall'] = imputer.transform(X_test[['Rainfall']])

# Create a RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=200, max_depth=5, min_samples_leaf= 20, random_state=42)

# The model is trained using the imputed training data
classifier.fit(X_train, y_train)

# The model makes predictions using the imputed test data
y_prediction = classifier.predict(X_test)

# The model is evaluated using the accuracy score and the classification report
accuracy = accuracy_score(y_test, y_prediction)
print(f"Accuracy: {accuracy}")

# Display additional classification metrics
print(classification_report(y_test, y_prediction))
```
